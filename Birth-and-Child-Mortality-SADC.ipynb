# Birth and Child Mortality in SADC countries

```python
# import pandas

import pandas as pd
```

```python
#create a dataframe from the downloaded excel file

raw_unicef_data = pd.read_excel('Unicef_data2.xlsx')
df = raw_unicef_data.copy()
df
```

```python
# create a checkpoint and drop unnecesary rows

df_dropped = df.drop([0, 1])
df_dropped.head()
```

```python
#get the column names

df_dropped.columns
```

```python
#rename the columns

df_dropped.columns = ['Country', 'Indicator', '2019', '2018', '2017', '2016', '2015',
       '2014', '2013', '2012', '2011', '2010', '2009']

df_dropped.head()
```

```python
#values are in the 1000s, edit dataframe to show this

value = 2
year = 2019

for value in range (df_dropped.shape[0])[2::3]:
    for year in range (2009,2020):
        df_dropped.loc[value,str(year)] = df_dropped.loc[value,str(year)]*1000 
df_dropped
```

```python
#Create a checkpoint
df_checkpoint1 = df_dropped.copy()
```

# Create an estimate for surviving children under 5

Create a new row with tnformation on the estimated number of children born that live until 5

```python
years = df_checkpoint1.columns[2:]
survival_est=[]
countries = list(df_checkpoint1['Country'].unique())
pos = 0
indices=range(2,df_dropped.shape[0],3)


for country in countries:
    pos=countries.index(country)
    calculated = [countries[pos], 'Survival Estimate']
    
#the loop to create the angola list
    for year in years:
        n=indices[pos]
        survivors = df_checkpoint1.loc[n,str(year)]-(df_checkpoint1.loc[n+1,str(year)]+df_checkpoint1.loc[n+2,str(year)])
        calculated.append(survivors)
    survival_est.append(calculated)
survival_est
```

## Append to the dataframe

```python
df_checkpoint2 = df_checkpoint1.copy()
df_checkpoint2.head()
```

## convert the survival_est list to a dataframe with the same column names

```python
df_checkpoint2.columns
```

```python
#convert the survival_est list to a dataframe with the same column names

survival_est1=pd.DataFrame(survival_est, columns = ['Country', 'Indicator', '2019', '2018', '2017', '2016', '2015', '2014',
       '2013', '2012', '2011', '2010', '2009'])
print(survival_est1.head())
```

```python
df_checkpoint2=pd.concat([df_checkpoint2, survival_est1], sort= False)
df_checkpoint2.tail(20)
```

```python
#save to CSV

df_checkpoint2.to_csv('sadc-cleaned.csv', index = False)
```
